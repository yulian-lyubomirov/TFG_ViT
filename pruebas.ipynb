```python
import torch

from functions.train_test import train, test,train_kd
from functions.data_loader import get_data_loader,load_checkpoint
from models.vit_early_exit import ViT_early_exit 
from models.CNN_ViT_early_exit import CNN_ViT_early_exit
from functions.helpers import count_parameters
import functions as f
```

```python
torch.manual_seed(42)
```

```python
train_loader, test_loader = get_data_loader(
    80, 2, "datasets/cifar-100/cifar-100-python", download=True
)
```

```python
model = ViT_early_exit(
    image_size=32,
    dim=64,
    patch_size=4,
    num_classes=100,
    depth=6,
    heads=6,
    mlp_dim=256,
    dropout=0.1,
    early_exit=False
).to("cuda")
```

```python
train(
    model,
    train_loader,
    test_loader,
    epochs=300,
    learning_rate=0.001,
    device="cuda",
    weight_decay=0.0005,
    # save_path='save_model/cifar-100/vit_early_exit_big',
    # load_path='save_model/cifar-100/vit_base/best_model.pt'
)
```

```python
model_test = ViT_early_exit(
    image_size=32,
    dim=64,
    patch_size=4,
    num_classes=100,
    depth=6,
    heads=6,
    mlp_dim=256,
    dropout=0.1,
    early_exit=True
).to("cuda")
load_path='save_model/cifar-100/vit_early_exit_big/best_model.pt'
load_checkpoint(model_test,load_path)
```

```python
train_loader_1_example, test_loader_1_example = get_data_loader(
    1, 2, "datasets/cifar-100/cifar-100-python", download=True
)
```

```python
test(model_test,test_loader_1_example,'cuda')
```

```python
CNN_ViT_early_exit = CNN_ViT_early_exit(
    image_size=32,
    dim=64,
    patch_size=4,
    num_classes=100,
    depth=6,
    heads=6,
    mlp_dim=256,
    dropout=0.1,
    early_exit=False
).to("cuda")
```

```python
train(
    CNN_ViT_early_exit,
    train_loader,
    test_loader,
    epochs=300,
    learning_rate=0.001,
    device="cuda",
    weight_decay=0.0005,
    # save_path='save_model/cifar-100/CNN_ViT_early_exit',
    # load_path='save_model/cifar-100/vit_base/best_model.pt'
)
```
